**pandas의 시간날짜 관련 기능**

|Concept|설명|
|---|---|
|Date time|한개일 때: `Timestamp`, 여러개일 때:`TimedeltaIndex`|
|Time delta|한개일 때: `Timedelta`, 여러개일 때:`TimedeltaIndex`|
|Time span|빈도에 의한 time delta(좀 더 확장개념)|
|Date offsets|dateutil에서 가져온 기능; python과 numpy에는 없는 기능|

pandas의 시간날짜 관련 공식문서: https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html

````python
# seaborn으로 상관관계를 그릴 수 있다.
# 위의 scatter_matrix와 같다.
sns.pairplot(iris)
````

````python
# seaborn도 matplotlib으로 만들었다.
import matplotlib.pyplot as plt
plt.style.use('ggplot')
````

````python
# seaborn은 더 세분화해서 분석할 수 있다.
# matplotlib에서 subplot으로 그릴 수 있지만, 세부설정을 하는데 오래 걸린다.
# 자주 사용하는 기능을 seaborn으로 쉽게 사용할 수 있다.
sns.pairplot(iris, hue='species')
````

````python
# 상관관계를 분석하는 기능.
iris.corr()
````

````python
# 상관관계를 시각화, 색이 연할 수록 상관관계가 낮다.
# matplotlib으로 한땀한땀 그릴 수도 있지만, seaborn으로 쉽게 그릴 수 있다.
sns.heatmap(iris.corr())
````

````python
# 분석할 때 기본적이고, 자주 사용하는 기능을 seaborn은 다 만들어두었다.
sns.violinplot(x='petal_width', y='sepal_length', data=iris)
````

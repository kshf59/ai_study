### DBSCAN

```python
import mglearn
```

DBSCAN은 밀도 기반으로 데이터를 군집화한다.  
어느 데이터를 기준으로 반경 $\epsilon$내에 다른 데이터가 $n$개 이상 있으면 하나의 군집으로 인식하는 방식이다.  
$\epsilon$은 하이퍼파라미터 eps이고, $n$은 하이퍼파라미터 min_samples이다.

```python
mglearn.plot_dbscan.plot_dbscan()
```

```python
from sklearn.cluster import DBSCAN
```

```python
# KMeans는 몇개로 클러스터링할지 미리 정해야했다.
# DBSCAN은 몇개로 클러스터링할지 정하지 않아도 데이터 기반으로 찾는다.
# 둘 다 거리 기반이라는 공통점이 있다.
db = DBSCAN()
```

```python
# 음수는 중요하지 않다.
db.fit_predict(data.data)
```

```python
db.labels_
```

```python
import numpy as np
```

```python
# DBSCAN으로 몇개의 클러스터로 묶이는지 먼저 확인하고, KMeans의 하이퍼파라미터로 사용할 수 있다.
# 여기서는 3개로 클러스터링했으므로, KMeans에서 k를 3으로 학습하는 것이 적절할 것이다. 
np.unique(db.labels_)
```

## Dimension Reducion

### PCA

PCA(Principle Component Analysis, 주성분분석)은 차원을 줄이는 알고리즘이다.

먼저, pca는 데이터의 분포를 가장 잘 나타낼 수 있는 축(component)을 찾는다.  
그리고 해당 축으로 데이터를 사영(projection) 시킨다. (사영: 빛을 비췄을 때 그림자가 생기는 것과 같음)  
데이터는 주성분 축으로 사영시켰을 때 분산이 가장 크다. (분산이 클 때 전체데이터를 가장 잘 설명한다.)

다음으로, 첫번째 주성분 축과 직교하면서 데이터를 가장 잘 나타내는(사영했을 때 분산이 큰) 축이 두번째 주성분이다.  
각 주성분으로 사영시켜 저차원으로 변환한 데이터는 원래의 데이터를 잘 나타낸다.

```python
mglearn.plot_pca.plot_pca_illustration()
```

```python
from sklearn.decomposition import PCA
```

```python
# feature를 3개로 줄인다.
pcav = PCA(3)
```

```python
pcav.fit_transform(data.data)
```

> 수정 : 객체 지향에서 overriding
> 추가 : duck-typing
> 수정 : decorator

```python
class Base:
    def __init__(self):
        print('Base')
class A(Base):
    def __init__(self):
        print('A')
class B(Base):
    def __init__(self):
        print('B')
class C(Base):
    def __init__(self):
        print('C')
```

```python
c = C()
```

## Tidy Data

- Tidy Data란 깔끔한 데이터 (=분석하기 좋은 데이터) 이다.
- Tidy Data의 3가지 조건
    - 각 변수는 개별의 열(column)으로 존재한다.
    - 각 관측치는 행(row)를 구성한다.
    - 각 테이블은 단 하나의 관측기준에 의해서 조직된 데이터를 저장한다.
- Multi index처럼 복잡한 구조를 갖지 않는다.
- 모든 데이터를 DataFrame(2차원)을 기본으로 한다.

`melt는` 열을 변수(variable)로 만들고, 값을 value라는 새로운 열에 추가한다. 데이터를 분석하기 좋은 형태로 가공한다.

#### wide_to_long

**wide vs long**

- 데이터가 추가되면 column을 늘려야한다 => 설계를 다시해야한다. (설계가 잘못됨)
    - wide format이라고 한다.
- 데이터가 추가되면 row를 늘려야한다. => 올바름
    - long format이라고 한다.
    - 데이터 추가/삭제/관리 가 편하다.

wide_to_long: wide format을 long format으로 바꾼다. (=melt)

````python
weather["index"] = weather.index
````

````python
pd.wide_to_long(weather, "d", "index", "variable")
````

### groupby

````python
tips.groupby(["sex", "day"])
````

꺾은 괄호를 반환하면 for문에 넣을 수 있는 경우가 많다.  

````python
# i는 index, j는 data
for i,j in tips.groupby(["sex", "day"]):
    print(i)
````

`groupby`는 집계 메서드를 붙여서 사용한다.

````python
tips.groupby(["sex", "day"]).sum()
````

---

python에도 `groupby`가 있었다.

````python
from itertools import groupby
````

````python
t = groupby([1,1,2,1,2,3,4,4,4])
````

````python
# next 대신 list로 내용을 한번에 볼 수 있다.
tt = list(t)
````

````python
# 연속되는 값은 1개로 묶였다.
tt
````

`Sorted`는 값을 순서대로 정렬하는 함수이다.  
`key`옵션은 비교하는 기준을 의미한다.

````python
sorted([(1,2,3), (2,1,4)], key=lambda x:x[1]) # 정렬기준을 tuple의 1번째 값으로 함
````

`groupby`도 `key`옵션이 있다.

````python
t = groupby([(1,1), (1,2), (3,4), (4,4)], key=lambda x:x[0]) # 기준을 tuple의 0번째 값으로 함
tt = list(t)
tt
````

````python
t = groupby([(1,1), (1,2), (3,4), (4,4)], key=lambda x:x[1]) # 기준을 tuple의 1번째 값으로 함
tt = list(t)
tt
````

---

집계함수를 한번에 여러개 쓸 수 있다.

````python
tips.groupby(["sex", "day"]).mean() # 집계함수를 1개 사용할 때는 메소드 방식
````

````python
tips.groupby(["sex", "day"]).aggregate([np.mean, np.sum]) # 집계함수를 여러개 사용할 때는 aggregate
````

````python
tips.groupby(["sex", "day"]).agg(["mean", "sum"]) # aggregate는 agg로 단축표현을 쓸 수 있다. (완전 동일)
                                                  # 함수 이름을 문자열로 쓸 수도 있다.  
````

집계 결과를 인덱싱하는 방법은 동일하다.

````python
# dictionary 방식
tips.groupby(["sex", "day"]).aggregate([np.mean, np.sum])["tip"]
````

````python
# attribute 방식
tips.groupby(["sex", "day"]).aggregate([np.mean, np.sum]).tip
````

````python
# fancy indexing -> tip이 그대로 남아 있는 차이점이 있다.
tips.groupby(["sex", "day"]).aggregate([np.mean, np.sum])[["tip"]]
````

인덱싱을 집계함수 쓰기 전에 인덱싱할 수도 있다. 

````python
# 앞에 쓰는 것이 성능이 더 좋다. (집계할 내용이 줄어든다.)
# 단, 앞에 쓰면 실수할 가능성이 높아진다.
tips.groupby(["sex", "day"])[["tip"]].aggregate([np.mean, np.sum])
````

`groupyby`로 분석한 결과를 출력하기도 용이하다.

````python
# 시각화
tips.groupby(["sex", "day"]).agg(["mean", "sum"])['tip'].plot.bar()
````

````python
# html 코드로 출력
print(tips.groupby(["sex", "day"]).agg(["mean", "sum"])['tip'].to_html())
````

`transform`은 그룹으로 묶인 상태에서 사용할 수 있는 메소드이다.

````python
s = tips.groupby(["sex", "day"])["tip"]
````

````python
s.transform(lambda x:x**2)
````

`transfrom`의 기능은 `apply`와 같다.

````python
s.apply(lambda x:x**2) # apply는 그룹상태에서 사용 가능
                       # map은 Error
````

`group_keys`옵션은 apply를 쓸 때 인덱스를 표시할지 여부를 결정한다.

````python
tips.groupby(["sex", "day"], group_keys=False).apply(lambda x:x.sex)
````

````python
tips.groupby(["sex", "day"], group_keys=True).apply(lambda x:x.sex) # default : group_keys=True
````

### pivot_table

index와 column만 지정하면 알아서 데이터를 만들어준다.

````python
# aggfunc 기본값: mean 
tips.pivot_table(index='sex', columns=['day'])['size']
````

column 여러개도 쓸 수 있다.

````python
tips.pivot_table(index='sex', columns=['day', 'smoker'])['size']
````

index도 여러개 쓸 수 있다.

````python
tips.pivot_table(index=['sex', "smoker"], columns=['day'])['size']
````

`margins`는 각 행/열별로 평균을 내준다. 

````python
tips.pivot_table(index='sex', columns=['day'], margins=True)['size']
````

### crosstab

`crosstab`은 function방식으로 써야한다.

````python
# aggfunc 기본값: count
pd.crosstab(index=tips.sex, columns=tips.day)
````

column, index는 여러개 쓸 수 있다.

````python
pd.crosstab(index=tips.sex, columns=[tips.day, tips.smoker])
````

`margins`는 합계를 내준다. 

````python
pd.crosstab(index=tips.sex, columns=[tips.day, tips.smoker], margins=True)
````

## multi index

인덱스가 계층구조로 된 것을 말한다.

````python
tips.groupby(["sex", "day"], group_keys=False).agg('mean')
````

column이 계층구조로 쌓인 것도 multi index라고 한다.

````python
tips.groupby(["sex", "day"], group_keys=False).agg(['mean', 'sum'])
````

계층구조로 된 column도 내부 구조상 MultiIndex 객체로 되어 있다.

````python
tips.groupby(["sex", "day"], group_keys=False).agg(['mean', 'sum']).columns
````

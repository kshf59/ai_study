#### iterrows

`iterrows`는 행을 하나씩 꺼낸다.

````python
for i in tips.iterrows():
    print(i)
````

#### itertuples

`namedtuple`의 형태로 꺼내준다.

````python
# namedtuple
for i in tips.itertuples():
    print(i)
````

`namedtuple`이므로 attribute 방식도 사용 가능하다.

````python
for i in tips.itertuples():
    print(i.tip)
````

`namedtuple`은 descriptor로 setter가 막혀있다. 그래서 값을 바꿀 수 없다.

````python
# AttributeError: can't set attribute
for i in tips.itertuples():
    if i.day == "Sun":
        i.day = "일"
````

````python
# Error 없이 실행된다.
for i,j in tips.iterrows():
    j.tip = 100
````

#### apply

````python
num_tips
````

`apply`는 axis 개념이 있다.

````python
num_tips.apply(lambda x: sum(x)) # default : axis=0
````

````python
num_tips.apply(lambda x: print(x.tip), axis=1)
````

````python
num_tips.apply(lambda x:x.tip + x.total_bill, axis=1)
````

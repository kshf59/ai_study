### ColumnTransformer

ColumnTransformer:
1. Pipeline을 결합
2. Pipeline - 특정 컬럼을 선택해서 적용

```python
# ColumnTransformer은 전처리를 적용할 column을 지정할 수 있다.
from sklearn.compose import ColumnTransformer
```

`ColumnTransformer`는 전처리를 적용할 column 이름을 알아야한다.

```python
# select_dtypes는 원하는 dtype만 뽑는다.
# 숫자 dtype만 가져온다.
mpg.select_dtypes(['int64', 'float64'])
```

```python
# 숫자 dtype인 column의 이름만 가져온다.
mpg_numeric = mpg.select_dtypes(['int64', 'float64']).columns.values
mpg_numeric
```

```python
# mpg는 target이므로 제외한다.
mpg_numeric = mpg_numeric[1:]
```

```python
# object dtype인 column의 이름만 가져온다.
mpg_category = mpg.select_dtypes(['object']).columns.values
mpg_category
```

dtype에 따라 다르게 적용할 Pipeline을 따로 만든다.  
이 때, 알고리즘은 넣지 않는다.

```python
mpg_numeric_transformer = Pipeline([('imputer', SimpleImputer())])
```

```python
# 문자열은 평균을 구할 수 없기 때문에, SimpleImputer의 strategy를 'most_frequent'로 바꾼다.
# most_frequent 옵션은 빈도를 기반으로 하므로, 문자열 데이터도 사용할 수 있다.
mpg_category_transformer = Pipeline([('imputer', SimpleImputer(strategy='most_frequent')), 
                                     ('onehot', OneHotEncoder())])
```

따로 만든 Pipeline을 `ColumnTransformer`에 넣는다.

```python
# 리스트 안에 튜플 형태
# 튜플 형식 : (이름, pipeline, 적용할 컬럼 이름)
all_pre = ColumnTransformer([('num', mpg_numeric_transformer, mpg_numeric), 
                             ('cat', mpg_category_transformer, mpg_category)])
```

`ColumnTransformer`를 `PipeLine`에 넣는다.  
`ColumnTransformer`는 전처리만 한다. 알고리즘을 여기에 추가한다.

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
pipe = Pipeline([('all', all_pre), ('knn', KNeighborsRegressor())])
```

```python
pipe.fit(mpg.iloc[:, 1:], mpg.iloc[:, 0])
```

이번에는 `OneHotEncoder` 대신에 `LabelEncoder`를 사용해보자.

```python
from sklearn.preprocessing import LabelEncoder
```

```python
mpg_numeric_transformer = Pipeline([('imputer', SimpleImputer())])
```

```python
# 이번에는 strategy를 'constant'로 해보자.
# strategy='constant'일 때는 fill_value가 있어야 한다.
mpg_category_transformer = Pipeline([('imputer', SimpleImputer(strategy='constant', fill_value='usa')), 
                                       ('onehot', OneHotEncoder())])
```

```python
all_pre = ColumnTransformer([('num', mpg_numeric_transformer, mpg_numeric), 
                             ('cat', mpg_category_transformer, mpg_category)])
```

```python
pipe = Pipeline([('all', all_pre), ('knn', KNeighborsRegressor())])
```

```python
pipe.fit(mpg.iloc[:,1:], mpg.iloc[:,0])
```

### ColumnTransformer와 map

`map`과 연동할 수 있는지 살펴보자

```python
# horsepower는 NaN이 있어서 에러를 일으키므로 지운다.
mpg_no_horsepwer = mpg.copy().drop(columns=['horsepower'])
```

```python
# map은 callable 객체를 인자로 넣어야 한다.
# TypeError: 'ColumnTransformer' object is not callable
mpg_no_horsepwer.origin.map(temp)
```

```python
# ColumnTransformer 인스턴스는 callable이 아니다.
# 전처리를 적용할 때는 fit_transform 메소드를 이용한다.
temp.fit_transform(mpg_no_horsepwer)
```

callable인 `fit_transform`을 `map`에 넣는다.  
`map`은 Series에만 쓸 수 있다. temp 안에 있는 `OneHotEncoder`는 2차원(DataFrame)이 필요하다.  
shape 때문에 `map`을 쓸 수 없다.

```python
# Expected 2D array, got scalar array instead
mpg_no_horsepwer.origin.map(temp.fit_transform)
```

DataFrame에서는 `map`대신에 `apply`를 쓸 수 있다.  
`apply`는 axis 옵션에 따라서 행/열 별로 적용한다.  
내부적으로 Series(1차원)별로 적용하고 다시 합쳐서 2차원으로 만든다.  

```python
# Error
mpg_no_horsepwer.apply(temp.fit_transform, axis=1)
```

DataFrame에서는 `map`대신에 `applymap`을 쓸 수 있다.

### passthrough

`Pipeline`에서 전처리를 적용하지 않을 때는 passthrough를 이용한다.

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# horsepower에는 NaN이 있으므로 SimpleImputer가 필요하다.
mpg_numeric_transformer = Pipeline([('imputer', SimpleImputer())])
```

```python
# 문자열 column은 숫자형으로 인코딩이 필요하다.
mpg_category_transformer = Pipeline([('onehot', OneHotEncoder())])
```

```python
all_pre = ColumnTransformer([('num', mpg_numeric_transformer, mpg_numeric), 
                             ('cat', mpg_category_transformer, mpg_category)])
```

```python
pipe = Pipeline([('all', all_pre), ('scale', StandardScaler()), ('clf', KNeighborsRegressor())])
```

```python
# scale을 바꿔가면서 GridSearch한다.
# passthrough일 때는 scale을 적용하지 않고 넘어간다.
grid = GridSearchCV(pipe, {'scale':['passthrough', StandardScaler(), MinMaxScaler()]})
```

```python
grid.fit(mpg.iloc[:,1:], mpg.iloc[:,0])
```

```python
pd.DataFrame(grid.cv_results_).T
```

```python
# ValueError: 1D data passed to a transformer that expects 2D data. 
x.fit_transform(mpg)
```

```python
# column 이름으로 list로 묶으면 2차원으로 넘겨진다.
x = ColumnTransformer([('ss', StandardScaler(), ['mpg'])])
```

```python
x.fit_transform(mpg)
```

column의 이름 대신 숫자를 쓸 수 있다.  
이때 숫자는 column의 인덱스 번호가 된다.

```python
# 'mpg' 대신 0
x = ColumnTransformer([('ss', StandardScaler(), [0])])
```

```python
# 0번째 column에 StandardScaler가 적용되었다.
x.fit_transform(mpg)
```

```python
# 숫자 1으로 바꿈
x = ColumnTransformer([('ss', StandardScaler(), [1])])
```

```python
# 1번째 column에 StandardScaler가 적용되었다.
x.fit_transform(mpg)
```

```python
# column index를 여러개 넣을 수 있다.
x = ColumnTransformer([('ss', StandardScaler(), [0, 1])])
```

```python
# ValueError
# 위에서 사용한 column index가 입력 데이터의 column 수보다 크면 에러가 발생한다.
# cylinders column 1개만 사용했기 때문에 1번쨰 column이 없다.
x.fit_transform(mpg_no_origin[['cylinders']])
```

```python
# 입력 데이터의 column 수가 더 크기 때문에 실행된다.
# 0번째 column, 1번째 column에 StandardScaler가 적용되었다.
x.fit_transform(mpg_no_origin)
```

```python
# [-1, 0]로 바꿈
# 음수는 뒤에서부터 시작한다.
x = ColumnTransformer([('ss', StandardScaler(), [-1, 0])])
```

```python
# -1번째 column과 0번째 column에 StandardScaler를 적용했다.
x.fit_transform(mpg_no_origin)
```

```python
# 0번째, 1번째, 2번째 column
# 여러개 사용할 수 있다.
x = ColumnTransformer([('ss', StandardScaler(), [0,1,2])])
```

```python
import tensorflow as tf
tf.debugging.set_log_device_placement(True)
```

```python
from sklearn.datasets import load_wine
```

```python
data = load_wine()
```

```python
def create_model():
    model= tf.keras.Sequential()
    model.add(tf.keras.layers.Dense(16, activation='relu', input_shape=(13,)))
    model.add(tf.keras.layers.Dense(3, activation='softmax'))
    model.compile(loss = 'sparse_categorical_crossentropy', optimizer='adam', metrics=['acc'])
    return model
```

```python
kc = tf.keras.wrappers.scikit_learn.KerasClassifier(create_model)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(kc, data.data, data.target, cv=10)
```

```python
from sklearn.datasets import load_boston
```

```python
bs = load_boston()
```

```python
bs.data.shape
```

```python
def create_model2():
    model = tf.keras.models.Sequential()
    model.add(tf.keras.layers.Dense(16, activation='relu', input_shape=(13,)))
    model.add(tf.keras.layers.Dense(1))
    model.compile(loss='mse', optimizer='adam', metrics=['mse'])
    return model
```

```python
kc2 = tf.keras.wrappers.scikit_learn.KerasRegressor(create_model2)
```

```python
cross_val_score(kc2, bs.data, bs.target)
```

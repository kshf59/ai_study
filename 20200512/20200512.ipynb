```python
import pandas as pd
```

```python
movies = pd.read_table("ml-1m\movies.dat", sep='::', engine='python', header=None,
                      names=['MovieID','Title','Genres'])
```

```python
#movies.set_index(0, inplace=True)
movies
```

```python
users = pd.read_csv(r"ml-1m\users.dat", sep='::', engine='python', header=None,
                   names=['UserID','Gender','Age','Occupation','Zip-code'])
```

```python
ratings = pd.read_csv(r"ml-1m\ratings.dat", sep='::', engine='python', header=None,
                     names=['UserID','MovieID','Rating','Timestamp'])
```

```python
ratings
```

```python
data = ratings.merge(movies).merge(users)
```

```python
recommendation_data = data[['UserID','MovieID','Rating']].copy()
```

```python
recommendation_pivot = recommendation_data.pivot('UserID','MovieID','Rating')
```

```python
recommendation_data.set_index(['UserID','MovieID']).unstack()
```

```python
recommendation_pivot.fillna(0, inplace = True)
```

```python
recommendation_pivot
```

```python
def movie_seen(user_id):
    return recommendation_pivot.loc[user_id][recommendation_pivot.loc[user_id]>0]
```

```python
movie_seen(5)
```

```python
small_test = recommendation_pivot.T.iloc[:500,:500]
```

```python
small_test_corr = small_test.corr()
```

```python
small_test_corr.loc[5]
```

```python
def nearest_user(user_id, n):
    return small_test_corr.loc[user_id].nlargest(n+1)[1:]
```

```python
nearest_user(5,5)
```
